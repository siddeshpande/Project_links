```python
df = df.copy()
df['start_date'] = pd.to_datetime(df['start'].apply(lambda x: x['dateTime']))
df['end_date'] = pd.to_datetime(df['end'].apply(lambda x: x['dateTime']))
df['meeting_duration'] = df['end_date'] - df['start_date']

df['meeting_date'] = df['start_date'].dt.date
df['meeting_day'] = df['start_date'].dt.day_name()

def calculate_min_work_duration(row):
    extra_duration = timedelta()
    if ( (timedelta(hours=time(9, 0).hour, minutes=time(9, 0).minute) - timedelta(hours=pd.to_datetime(row['start_date']).time().hour, minutes=pd.to_datetime(row['start_date']).time().minute)) > timedelta(seconds=0)):
        if ( (timedelta(hours=pd.to_datetime(row['end_date']).time().hour, minutes=pd.to_datetime(row['end_date']).time().minute) - timedelta(hours=time(9, 0).hour, minutes=time(9, 0).minute)) > timedelta(seconds=0) ):
            extra_duration += ( timedelta(hours=time(9, 0).hour, minutes=time(9, 0).minute) - timedelta(hours=pd.to_datetime(row['start_date']).time().hour, minutes=pd.to_datetime(row['start_date']).time().minute) )
        else:
            extra_duration += ( timedelta(hours=pd.to_datetime(row['end_date']).time().hour, minutes=pd.to_datetime(row['end_date']).time().minute) - timedelta(hours=pd.to_datetime(row['start_date']).time().hour, minutes=pd.to_datetime(row['start_date']).time().minute) )

    if ( (timedelta(hours=pd.to_datetime(row['end_date']).time().hour, minutes=pd.to_datetime(row['end_date']).time().minute) - timedelta(hours=time(17, 0).hour, minutes=time(17, 0).minute)) >= timedelta(seconds=0)):
        if ( (timedelta(hours=pd.to_datetime(row['start_date']).time().hour, minutes=pd.to_datetime(row['start_date']).time().minute) - timedelta(hours=time(17, 0).hour, minutes=time(17, 0).minute)) > timedelta(seconds=0) ):
            extra_duration += ( timedelta(hours=pd.to_datetime(row['end_date']).time().hour, minutes=pd.to_datetime(row['end_date']).time().minute) - timedelta(hours=pd.to_datetime(row['start_date']).time().hour, minutes=pd.to_datetime(row['start_date']).time().minute) )
        else:
            extra_duration += ( timedelta(hours=pd.to_datetime(row['end_date']).time().hour, minutes=pd.to_datetime(row['end_date']).time().minute) - timedelta(hours=time(17, 0).hour, minutes=time(17, 0).minute) )

    return extra_duration

df['extra_work_hours'] = df.apply(calculate_min_work_duration, axis=1)

df_2 = pd.DataFrame({'meeting_day': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']})

merged_df = pd.merge(df, df_2, on='meeting_day', how='right')
df_meet_time = merged_df.groupby('meeting_day').agg({'meeting_duration': 'sum',
                                            'extra_work_hours': 'sum'
                                            }).reset_index()

df_meet_time
```

```python
def calculate_meetings(unique_attendees_dict, start_time, summary, attendees):
    attendee_email_set = set()
    for attendee_dict in attendees:
        if 'self' not in attendee_dict and attendee_dict['responseStatus']!='declined':
            attendee_email_set.add(attendee_dict['email'])

    sorted_emails = tuple(sorted(attendee_email_set))
    if sorted_emails in unique_attendees_dict.keys():
        unique_attendees_dict[sorted_emails]['count'] += 1
    else:
        unique_attendees_dict[sorted_emails] = {'count': 1, 'summary': summary, 'meet_time': start_time}

    return unique_attendees_dict
```

```python
def calculate_cog_div(percentiles):
    ide_per = []
    emo_per = []
    act_per = []
    ord_per = []
    outlier = []

    ide_per.append(owner_percentiles['ideas']/100)
    emo_per.append(owner_percentiles['relational']/100)
    act_per.append(owner_percentiles['action']/100)
    ord_per.append(owner_percentiles['order']/100)

    for person in percentiles:
        if (len(person)>0):
            ide_per.append(person[0][0]/100)
            emo_per.append(person[0][1]/100)
            act_per.append(person[0][2]/100)
            ord_per.append(person[0][3]/100)

    ide_iqr = np.percentile(ide_per, 75) - np.percentile(ide_per, 25)
    emo_iqr = np.percentile(emo_per, 75) - np.percentile(emo_per, 25)
    act_iqr = np.percentile(act_per, 75) - np.percentile(act_per, 25)
    ord_iqr = np.percentile(ord_per, 75) - np.percentile(ord_per, 25)

    cog_div = round((ide_iqr+emo_iqr+act_iqr+ord_iqr)/4.0, 3)

    high_outlier = [(1.5*ide_iqr)+np.median(ide_per), (1.5*emo_iqr)+np.median(emo_per), (1.5*act_iqr)+np.median(act_per), (1.5*ord_iqr)+np.median(ord_per)]
    low_outlier = [np.median(ide_per)-(1.5*ide_iqr), np.median(emo_per)-(1.5*emo_iqr), np.median(act_per)-(1.5*act_iqr), np.median(ord_per)-(1.5*ord_iqr)]

    if (owner_percentiles['ideas']/100>high_outlier[0] or owner_percentiles['ideas']/100<low_outlier[0]):
        outlier.append('Ideas')
    if (owner_percentiles['relational']/100>high_outlier[1] or owner_percentiles['relational']/100<low_outlier[1]):
        outlier.append('Relational')
    if (owner_percentiles['action']/100>high_outlier[2] or owner_percentiles['action']/100<low_outlier[2]):
        outlier.append('Action')
    if (owner_percentiles['order']/100>high_outlier[3] or owner_percentiles['order']/100<low_outlier[3]):
        outlier.append('Order')

    return cog_div, outlier
```

```python
cog_div_list = []
outlier_list = []
for index, row in df_group.iterrows():
    cog_outlier_scores = calculate_cog_div(row['Percentiles'])
    cog_div_list.append(cog_outlier_scores[0])
    outlier_list.append(cog_outlier_scores[1])

df_group['Cognitive Diversity scores'] = cog_div_list
df_group['Outlier'] = outlier_list
```

```python
import random

# Adjectives for different styles
adjectives = {
    'ideas': {
        5: ["leadership roles", "visionary", "creative", "practical"],
        4: ["strategic", "insightful", "analytical"],
        3: ["systems-oriented", "implementing", "pragmatic", "balanced"],
        2: ["practical", "build processes", "organized", "detail-oriented", "solution-oriented"],
        1: ["process-oriented", "documentation-oriented", "efficient", "data-driven"]
    },
    'relational': {
        5: ["friend", "emotionally intelligent", "supportive", "relationship-oriented"],
        4: ["collaborative", "supportive", "encouraging", "empathetic"],
        3: ["emotionally balanced", "rational", "conflict-resolution", "diplomatic"],
        2: ["authentic", "straightforward", "direct", "honest", "objective"],
        1: ["no friends", "pragmatic", "logical", "fact-based", "problem-solving"]
    },
    'action': {
        5: ["leadership roles", "authority", "expedient", "decisive", "motivating"],
        4: ["influential", "commitment", "empowering", "strategic"],
        3: ["shared-authority", "collaborative", "team-oriented"],
        2: ["process-oriented", "goal-oriented", "strategic"],
        1: ["methodical", "need guidance", "team-oriented", "action-oriented"]
    },
    'order': {
        5: ["structured", "perfection", "controlling", "detail-oriented", "organized", "predictable"],
        4: ["orderly", "enabler", "efficient", "results-oriented", "detail-oriented"],
        3: ["creates efficient systems", "balanced", "adaptable", "systems-oriented"],
        2: ["delegates responsibility", "adaptable", "flexible", "hands-off", "creative"],
        1: ["free-spirited", "creative", "very flexible", "independent"]
    }
}

def pick_random_trait(scores):
    random_trait_focus = random.choice(list(scores.items()))
    scale = random_trait_focus[0]
    score = random_trait_focus[1]
    if score in adjectives[scale]:
        member_adj = ', '.join(adjectives[scale][score])
    owner_score = final_stats_df.loc[0,'owner_wethos_score'][scale]
    owner_adj = ', '.join(adjectives[scale][owner_score])
    return member_adj, owner_adj

```
